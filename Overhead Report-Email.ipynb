```python
#--------------------------------Top 12 files--------------------------
import pandas as pd
from openpyxl import load_workbook
import os

# File paths
source_template = r"C:\Users\141823\OneDrive - Etex Group\Desktop\Etex AU - Finance - Finance\19.Business Systems\Overhead Report\Etex Consolidated- Overhead Report-Template.xlsm"
reference_file = r"C:\Users\141823\OneDrive - Etex Group\Desktop\Etex AU - Finance - Finance\19.Business Systems\Overhead Report\Email List.xlsx"
output_folder = r"C:\Users\141823\OneDrive - Etex Group\Desktop\Etex AU - Finance - Finance\19.Business Systems\Overhead Report\Send\Review"

# Load the reference mapping
df_ref = pd.read_excel(reference_file, sheet_name=0, usecols="B:C", skiprows=1, names=["Workbook", "Sheet Name"])

# Group the sheet names by workbook name
sheet_groups = df_ref.groupby("Workbook")["Sheet Name"].apply(list).to_dict()

# Loop through each workbook and its corresponding sheet list
for workbook_name, sheet_list in sheet_groups.items():
    # Open the workbook you want to paste into
    dest_path = os.path.join(output_folder, workbook_name)
    if not os.path.exists(dest_path):
        print(f"Workbook not found: {dest_path}")
        continue

    dest_wb = load_workbook(dest_path)

    # Load source sheets from template
    source_wb = load_workbook(source_template, data_only=True)

    for sheet_name in sheet_list:
        if sheet_name not in source_wb.sheetnames:
            print(f"Sheet '{sheet_name}' not found in source template.")
            continue

        source_ws = source_wb[sheet_name]

        if sheet_name == "Transaction":
            # Remove old sheet and copy from source
            if sheet_name in dest_wb.sheetnames:
                del dest_wb[sheet_name]
            dest_wb.create_sheet(sheet_name)
            dest_ws = dest_wb[sheet_name]

            for row in source_ws.iter_rows():
                for cell in row:
                    dest_ws[cell.coordinate].value = cell.value

        else:
            # Replace only D6:AA188
            if sheet_name not in dest_wb.sheetnames:
                print(f"Sheet '{sheet_name}' not found in destination workbook.")
                continue

            dest_ws = dest_wb[sheet_name]
            for i, row in enumerate(source_ws.iter_rows(min_row=6, max_row=188, min_col=4, max_col=27), start=6):
                for j, cell in enumerate(row, start=4):
                    dest_ws_cell = dest_ws.cell(row=i, column=j)
                    dest_ws_cell.value = cell.value  # copy value only (no formula) 
                    # Do NOT touch style — this keeps destination formatting intact

    # Save changes
    dest_wb.save(dest_path)
 
```

```python
import pandas as pd
import os
import win32com.client as win32

## Your paths
folder_path = r"C:\Users\141823\OneDrive - Etex Group\Desktop\Etex AU - Finance - Finance\19.Business Systems\Overhead Report\Send\Review"
reference_file = r"C:\Users\141823\OneDrive - Etex Group\Desktop\Etex AU - Finance - Finance\19.Business Systems\Overhead Report\Email List.xlsx"

# 1. Load your email reference
df_email = pd.read_excel(reference_file, sheet_name='Top Email')
df_email.columns = df_email.columns.str.strip()  # Clean column names




```

```python
# 3. Launch Outlook
outlook = win32.Dispatch('outlook.application')

# 4. Loop through each row
for index, row in df_email.iterrows():
    file_to_attach = os.path.join(folder_path, row['File'])
    
    if not os.path.exists(file_to_attach):
        print(f"File not found: {file_to_attach}")
        continue

    # Create a new email
    mail = outlook.CreateItem(0)  # 0 = olMailItem
    mail.To = row['To']
    mail.CC = row['CC']
    mail.Subject = f"Overhead Report - {row['Name']}"
    mail.Body = f"""Hi {row['Name']},

Please find attached the Overhead Report for your review.

Let me know if you have any questions.

Thanks,
Lee Zheng
"""
    # Attach the file
    mail.Attachments.Add(file_to_attach)

    # Save as draft (or you can use mail.Send() if you want to auto-send)
    mail.Save()
```
